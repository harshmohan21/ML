```python
import pickle
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
import numpy as np

# Load the pickle file
with open('./data/mnist_small.pkl', 'rb') as f:
    data = pickle.load(f)

X = data['X']
Y = data['Y']

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

# Visualize PCA results using Seaborn
for i in range(10):
    indices = np.where(Y == i)
    sns.scatterplot(x=pca_result[indices, 0].flatten(), y=pca_result[indices, 1].flatten(), label=str(i), marker='.')
plt.title('PCA')
plt.legend()

plt.tight_layout()
# plt.savefig('part3_PCA.png')
plt.show()
```

```python
# Perform t-SNE
tsne = TSNE(n_components=2)
tsne_result = tsne.fit_transform(X)

# Visualize t-SNE results using Seaborn
for i in range(10):
    indices = np.where(Y == i)
    sns.scatterplot(x=tsne_result[indices, 0].flatten(), y=tsne_result[indices, 1].flatten(), label=str(i), marker='.')
plt.title('t-SNE')
plt.legend()

plt.tight_layout()
# plt.savefig('part3_t-SNE.png')
plt.show()
```
